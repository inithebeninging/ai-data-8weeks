```python
# Problem 1:introduce oneslef
# 여기에 코드 작성
name = "inithebeninging"
age = "10"
major = "electronic engineering"
print("Hi, my name is", name, "and I am", age, "years old. My major is", major, ".")
```

```python
# Problem 2:basic mathematics
# 여기에 코드 작성
a = 15
b = 7
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
```

```python
# Problem 3:Fahrenheit to Celcius
# 여기에 코드 작성
c = 25
F = c * 9/5 + 32
print(F)
```

```python
# Problem 4:list making
# 여기에 코드 작성
scores = [67, 69, 90, 100, 99]
print("first score", scores[0])
print("last one", scores[-1])
print("numbers of score", len(scores))
print("total", sum(scores))
print("average score", sum(scores)/len(scores))
```

```python
# Problem 5:using append
# 여기에 코드 작성
todo = [] #make a empty list
a = "workout"
b = "reading"
c = "working"
todo.append(a) #you can add data into list by usin listname.append(data)
todo.append(b)
todo.append(c)
print(todo)
```

```python
# Problem 6:using for
# 여기에 코드 작성
subjects = ["Python", "SQL", "Machine Learning", "English"]
for s in subjects:
    print("I will study", s) #리스트 안에 있는 요소들의 명칭을 내가 지정가능
```

```python
for n in range(1,11): #range(start,stop)Generates numbers from start up to stop - 1
  print(n)
```

```python
# Problem 8:add one to hundred
# 여기에 코드 작성
total = 0 #항상 초기화하기

for i in range(101):
  total += i

print(total)
```

```python
# Problem 9:even number
# 여기에 코드 작성
for a in range(1, 21):
  if a % 2 == 0:
    print(a)
```

```python
# Problem 10:over 80
# 여기에 코드 작성
scores = [55, 70, 82, 90, 67, 100, 45]
for a in range(0,7):
  if scores[a]>= 80:
    print(scores[a])
```

```python
# Problem 11: dictionary
# 여기에 코드 작성
student = {
    "name": "pororo",
    "age": 5,
    "major": "law"
}
print(student["name"])#print by key factor
```

```python
# Problem 12:add value in dictionary
# 여기에 코드 작성
student = {
    "name": "pororo",
    "age": 5,
    "major": "law"
}
student["grade"]=3 #add
print(student) #i shall remember how i do about dictionary
```

```python
# Problem 13:to make a function
# 여기에 코드 작성
def fun(a, b):
  sum = a + b
  return sum

result = fun(3, 5)
print(result)
```

```python
# Problem 14:to make a function
# 여기에 코드 작성
def check(a):
  if a > 60:
    print("pass")
  else:
    print("fail")

check(50)
```

```python
# Problem 15:avg, pass
# 여기에 코드 작성
def check(a):
  if a > 80:
    print("pass")
  else:
    print("fail")

marks = [70, 85, 90, 55, 100]
for a in range(0,5):
  check(marks[a])
average = sum(marks) / len(marks)
print(average)
```
