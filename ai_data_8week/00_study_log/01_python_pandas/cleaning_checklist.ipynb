# Day 4 정리

## 오늘 배운 것

- 결측치는 비어 있는 값이다.
- isnull().sum()으로 컬럼별 결측치 개수를 확인할 수 있다.
- duplicated().sum()으로 중복 행 개수를 확인할 수 있다.
- dtypes로 컬럼별 자료형을 확인할 수 있다.
- astype()은 자료형을 직접 바꿀 때 사용한다.
- to_numeric()은 문자형 숫자를 숫자형으로 바꿀 때 사용한다.
- errors="coerce"를 사용하면 변환할 수 없는 값을 NaN으로 바꿀 수 있다.
