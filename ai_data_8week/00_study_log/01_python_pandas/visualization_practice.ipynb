## 오늘 목표
학생 예제 데이터를 사용해 그래프 5개를 그리고, 각 그래프에 대한 해석 문장을 작성한다.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "name": ["Jisoo", "Minho", "Yuna", "David", "Sara", "Tom", "Hana", "John", "Mina", "Chris",
             "Nari", "Leo", "Sumin", "Kevin", "Jane"],
    "gender": ["F", "M", "F", "M", "F", "M", "F", "M", "F", "M",
               "F", "M", "F", "M", "F"],
    "grade": [1, 2, 3, 1, 2, 3, 1, 4, 4, 2,
              3, 4, 2, 1, 3],
    "major": ["EE", "CS", "DS", "EE", "CS", "DS", "EE", "CS", "DS", "EE",
              "CS", "DS", "EE", "CS", "DS"],
    "score": [82, 75, 91, 68, 88, 79, 95, 84, 90, 72,
              86, 77, 93, 65, 89],
    "study_hours": [5, 3, 6, 2, 5, 4, 7, 4, 6, 3,
                    5, 4, 7, 2, 6],
    "pass": ["Pass", "Pass", "Pass", "Fail", "Pass", "Pass", "Pass", "Pass", "Pass", "Pass",
             "Pass", "Pass", "Pass", "Fail", "Pass"]
}

df = pd.DataFrame(data)
df.head()
df.shape
df.info()
df.describe()
```

```python
gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Number of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## Graph 1 해석

이 막대그래프는 성별 학생 수를 보여준다.  
F 학생 수가 M 학생 수보다 조금 더 많다.

```python
grade_avg_score = df.groupby("grade")["score"].mean().sort_index()

plt.figure(figsize=(6, 4))
plt.plot(grade_avg_score.index, grade_avg_score.values, marker="o")
plt.title("Average Score by Grade")
plt.xlabel("Grade")
plt.ylabel("Average Score")
plt.xticks(grade_avg_score.index)
plt.show()
```

## Graph 2 해석

이 선그래프는 학년별 평균 점수의 변화를 보여준다.  
이 데이터에서는 2학년과 3학년의 평균 점수가 상대적으로 높고, 1학년은 낮은 편이다.

```python
plt.figure(figsize=(6, 4))
plt.hist(df["score"], bins=5)
plt.title("Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

## Graph 3 해석

이 히스토그램은 학생 점수 분포를 보여준다.  
대부분의 점수가 70점 이상에 분포하고 있으며, 60점대 점수는 상대적으로 적다.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df["study_hours"], df["score"])
plt.title("Study Hours vs Score")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.show()
```

## Graph 4 해석

이 산점도는 공부 시간과 점수의 관계를 보여준다.  
대체로 공부 시간이 많을수록 점수가 높아지는 경향이 보인다.

## 오늘 배운 것

- bar chart는 범주별 개수나 평균을 비교할 때 사용한다.

- 그래프를 그린 뒤에는 반드시 해석 문장을 붙여야 한다.
