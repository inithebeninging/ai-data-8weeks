# Day 3 정리

## 오늘 배운 것

- read_csv()는 CSV 파일을 불러올 때 사용한다.
- head()는 데이터 앞부분을 확인할 때 사용한다.
- shape는 데이터의 행과 열 개수를 확인할 때 사용한다.
- info()는 컬럼별 데이터 개수와 자료형을 확인할 때 사용한다.
- describe()는 숫자형 데이터의 기초 통계를 확인할 때 사용한다.
- isnull().sum()은 결측치 개수를 확인할 때 사용한다.
